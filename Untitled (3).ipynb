```python
from tkinter import *
from PIL import ImageTk, Image
import turtle


root = Tk()
root.geometry("950x500")
root.title("SAFE BET")

iconImg =ImageTk.PhotoImage(Image.open("logo.jpg")) 
img = ImageTk.PhotoImage(Image.open("logo.jpg"))
continuebuttonImg = ImageTk.PhotoImage(Image.open("continuebutton.jpg"))
donebuttonImg = ImageTk.PhotoImage(Image.open("donebutton.jpg"))
exitbuttonImg = ImageTk.PhotoImage(Image.open("exitbutton.jpg"))

root.iconphoto(False, iconImg)
root.config(bg = "#1b0b54")


con_exit_frame = Frame(root, bg = "#1b0b54")
con_exit_frame.pack(side = BOTTOM)


def draw_grid(r,c,mirrorm,mirrorn,forwardPath,backwardPath):
    
    
    NUM_BOXES = r*c
    BOX_SIZE_FRACTION = 1/10
    GRID_SIZE  = 400
    BOX_SIZE = int(GRID_SIZE * BOX_SIZE_FRACTION)
    
    canvas = Canvas(root,bg="#1b0b54",width=GRID_SIZE+100, height=GRID_SIZE+100, highlightbackground= "#1b0b54")
    canvas.pack(padx = 0,pady = 10)
    
    t = turtle.RawTurtle(canvas)

    t.penup()
    t.goto(-GRID_SIZE/2 + BOX_SIZE/2, GRID_SIZE/2 - BOX_SIZE/2)
    t.pendown()
    t.speed(40)

    for i in range(r):
        for j in range(c):
            for k in range(4):
                t.forward(BOX_SIZE)
                t.right(90)
            
 
            
            
            t.penup()
            t.forward(BOX_SIZE)
            t.pendown()
        t.penup()
        t.goto(-GRID_SIZE/2 + BOX_SIZE/2, GRID_SIZE/2 - (i+2)*BOX_SIZE + BOX_SIZE/2)
        t.pendown()
    
    

    for i in range(r):
        for j in range(c):
            if [j+1, i+1] in mirrorn:
                t.penup()
                t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + i*BOX_SIZE, (GRID_SIZE/2 - BOX_SIZE/2) - j*BOX_SIZE)
                t.pendown()
                t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + (i+1)*BOX_SIZE, (GRID_SIZE/2 - BOX_SIZE/2) - (j+1)*BOX_SIZE)
                t.penup()
            elif [j+1,i+1] in mirrorm:
                t.penup()
                t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + (i)*BOX_SIZE  , (GRID_SIZE/2 - BOX_SIZE/2) - (j+1)*BOX_SIZE)
                t.pendown()
                t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + (i+1)*BOX_SIZE, (GRID_SIZE/2 - BOX_SIZE/2) - (j)*BOX_SIZE )
                t.penup()

        
        t.penup()
        t.goto(-GRID_SIZE/2 + BOX_SIZE/2, GRID_SIZE/2 - (i+2)*BOX_SIZE + BOX_SIZE/2)
        t.pendown()

        
    t.speed(5)
    t.penup()
    t.pencolor('red')
    t.pensize(3)
    t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + (forwardPath[0][1]-1)*BOX_SIZE+BOX_SIZE/2, (GRID_SIZE/2 - BOX_SIZE/2) - (forwardPath[0][0]-1)*BOX_SIZE-BOX_SIZE/2)
    t.pendown()
    for coord in forwardPath[1:]:
        y, x = coord
        x_offset = (-GRID_SIZE/2 + BOX_SIZE/2) + (x-1)*BOX_SIZE +BOX_SIZE/2
        y_offset = (GRID_SIZE/2 - BOX_SIZE/2) - (y-1)*BOX_SIZE-BOX_SIZE/2
        t.goto(x_offset, y_offset)

    if backwardPath != []:
        t.penup()
        t.pencolor('green')
        t.pensize(3)
        t.goto((-GRID_SIZE/2 + BOX_SIZE/2) + (backwardPath[0][1]-1)*BOX_SIZE+BOX_SIZE/2, (GRID_SIZE/2 - BOX_SIZE/2) - (backwardPath[0][0]-1)*BOX_SIZE-BOX_SIZE/2)
        t.pendown()
        for coord in backwardPath[1:]:
            y, x = coord
            x_offset = (-GRID_SIZE/2 + BOX_SIZE/2) + (x-1)*BOX_SIZE +BOX_SIZE/2
            y_offset = (GRID_SIZE/2 - BOX_SIZE/2) - (y-1)*BOX_SIZE-BOX_SIZE/2
            t.goto(x_offset, y_offset)




def changing_direction(direction, typeOfMirror):
  if typeOfMirror == 'm':
    if direction == 'horizontallyForward':
      nextDirection = 'verticallyUp'
    elif direction == 'horizontallyBackward':
      nextDirection = 'verticallyDown'
    elif direction == 'verticallyDown':
      nextDirection = 'horizontallyBackward'
    elif direction == 'verticallyUp':
      nextDirection = 'horizontallyForward'
  elif typeOfMirror == 'n':
    if direction == 'horizontallyForward':
      nextDirection = 'verticallyDown'
    elif direction == 'horizontallyBackward':
      nextDirection = 'verticallyUp'
    elif direction == 'verticallyDown':
      nextDirection = 'horizontallyForward'
    elif direction == 'verticallyUp':
      nextDirection = 'horizontallyBackward'
  return nextDirection



def moving_up(pos,mirrorn, mirrorm, r, c):
  i,j = pos[0]
  nextPos = []
  intialDirection = pos[1]
  for k in mirrorm+mirrorn:
    if k[1] == j and k[0] < i:
      nextPos = k
      break
  if nextPos in mirrorm:
    typeOfMirror = 'm'
  else:
    typeOfMirror = 'n'
  if nextPos == []:
    nextPos = [1,j]
  directionChanged = changing_direction(pos[1], typeOfMirror)
  return [nextPos, directionChanged, typeOfMirror,  intialDirection ]



def moving_down(pos,mirrorn, mirrorm, r, c):
  i,j = pos[0]
  nextPos = []
  intialDirection = pos[1]
  for k in mirrorm+mirrorn:
    if k[1] == j and k[0] > i:
      nextPos = k
      break
  if nextPos in mirrorm:
    typeOfMirror = 'm'
  else:
    typeOfMirror = 'n'
  if nextPos == []:
    nextPos = [r,j]
  directionChanged = changing_direction(pos[1], typeOfMirror)
  return [nextPos, directionChanged, typeOfMirror,  intialDirection, ]


def moving_forward(pos,mirrorn, mirrorm, r, c):
  i,j = pos[0]
  nextPos = []
  intialDirection = pos[1]
  for k in mirrorm+mirrorn:
    if k[0] == i and k[1] > j:
      nextPos = k
      break
  if nextPos in mirrorm:
    typeOfMirror = 'm'
  else:
    typeOfMirror = 'n'
  if nextPos == []:
    nextPos = [i,c]
  directionChanged = changing_direction(pos[1], typeOfMirror)
  return [nextPos, directionChanged, typeOfMirror, intialDirection ]



def moving_backward(pos,mirrorn, mirrorm, r, c):
  i,j = pos[0]
  nextPos = []
  intialDirection = pos[1]
  for k in mirrorm+mirrorn:
    if k[0] == i and k[1] < j:
      nextPos = k
      break
  if nextPos in mirrorm:
    typeOfMirror = 'm'
  else:
    typeOfMirror = 'n'
  if nextPos == []:
    nextPos = [i,1]
  directionChanged = changing_direction(pos[1], typeOfMirror)
  return [nextPos, directionChanged,typeOfMirror , intialDirection]



def tracking_path(path,mirrorn, mirrorm, r, c):
  pos = path[-1]
  while pos[0] in mirrorm+mirrorn and pos[0] != [r,c]:
    if pos[1] == 'horizontallyForward':
      pos = moving_forward(pos,mirrorn, mirrorm, r, c)
      path.append(pos)
    elif pos[1] == 'horizontallyBackward':
      pos = moving_backward(pos, mirrorn, mirrorm, r, c)        
      path.append(pos)
    elif pos[1] == 'verticallyUp':
      pos = moving_up(pos, mirrorn, mirrorm, r, c)
      path.append(pos)
    elif pos[1] == 'verticallyDown':
      pos = moving_down(pos, mirrorn, mirrorm, r, c)
      path.append(pos)

  return path



def forward_tracing(mirrorn, mirrorm, r, c):
  intial = [[1,1], "horizontallyForward"]
  pos = moving_forward(intial,mirrorn, mirrorm, r, c)
  forwardPath = [] 
  forwardPath.append(intial)
  forwardPath.append(pos)
  forwardPath = tracking_path(forwardPath,mirrorn, mirrorm, r, c)
  return forwardPath



def back_tracing(mirrorn, mirrorm, r, c):
  intial = [[r,c], "horizontallyBackward"]
  pos = moving_backward(intial,mirrorn, mirrorm, r, c)
  backwardPath = []
  backwardPath.append(intial)
  backwardPath.append(pos)
  backwardPath = tracking_path(backwardPath, mirrorn, mirrorm, r, c)
  return backwardPath



def finding_points(path, mirrorm, mirrorn):
  pointInPath = []
  
  for i in range(0, len(path)-1):
    [intial, final] = path[i:i+2]
    if intial[1] == "horizontallyForward":
      pointInPath.append(intial[0])
      while intial[0][1] < final[0][1]:
        intial[0] = [intial[0][0], intial[0][1]+1]
        
        pointInPath.append(intial[0])
       
    
    elif intial[1] == "verticallyDown":
      pointInPath.append(intial[0])
      while intial[0][0] < final[0][0]:
        intial[0] = [intial[0][0]+1, intial[0][1]]
        
        pointInPath.append(intial[0])
        

    elif intial[1] == "horizontallyBackward":
      pointInPath.append(intial[0])
      while intial[0][1] > final[0][1]:
        intial[0] = [intial[0][0], intial[0][1]-1]
        
        pointInPath.append(intial[0])
        

    elif intial[1] == "verticallyUp":
      pointInPath.append(intial[0])
      while intial[0][0] > final[0][0]:
        intial[0] = [intial[0][0]-1, intial[0][1]]
        
        pointInPath.append(intial[0])
        

  return pointInPath



def point_of_intersections(forwardPath, backwardPath,mirrorm,mirrorn): 
  pointsInforwardPath = finding_points(forwardPath,mirrorm,mirrorn)
  pointsInBackwardPath = finding_points(backwardPath,mirrorm,mirrorn)

  intersections = []
  for i in pointsInforwardPath:
    for j in pointsInBackwardPath:
      if i== j:
        intersections.append(i)
  return [intersections, pointsInforwardPath, pointsInBackwardPath]




def safe_bet(r,c,m,n,mirrormstr, mirrornstr):
    
    
    mirrorm = []
    if mirrormstr == ['[]']:
        mirrorm = []
    else:
        for i in mirrormstr:
             coordinates = list(i[1:len(i)-1].split(","))
             mirrorm.append(list(map(int, coordinates)))
    
    
    mirrorn = []
    if mirrornstr == ['[]']:
        mirrorn = []
    else:
        for j in mirrornstr:
            coordinates = list(j[1:len(j)-1].split(","))
            mirrorn.append(list(map(int, coordinates)))

          
    forwardPath = forward_tracing(mirrorn, mirrorm, r, c)
    
    
    if(forwardPath[-1][0] == [r,c]):
      forwardpathpoints = finding_points(forwardPath,mirrorm,mirrorn)
      draw_grid(r,c,mirrorm,mirrorn, forwardpathpoints,[])
      finalConclusion = "Your Safe is NOT SECURE"
      
       
    else:
      backwardPath = back_tracing(mirrorn, mirrorm, r, c)
      [intersections, pointsInforwardPath, pointsInBackwardPath]  = point_of_intersections(forwardPath, backwardPath,mirrorm,mirrorn)
      
      
      if intersections == []:
        finalConclusion = "No intersection points so Your Safe cannot be opened"
      else:
        finalConclusion = ("Your Safe is secure you have {} possible coordinates \n and least one is [{}, {}]".format(len(intersections), intersections[0][0], intersections[0][1]))
      
      draw_grid(r,c,mirrorm,mirrorn, pointsInforwardPath,pointsInBackwardPath)

    finalMessage = Label(con_exit_frame, text = finalConclusion,fg = "#ffffff", bg = "#1b0b54",font=("Alegreya", 15) )
    finalMessage.pack()
    
    
    
    

def secondWindow():
    main_frame.destroy()
    main_canvas.destroy()
    continue_btn.destroy()
    
    
    def thirdWindow():
        [r,c] = list(map(int,rowAndColumnEntry.get().split("x")))
        [m,n] = list(map(int,mAndnMirrorsEntry.get().split(",")))
        mirrormstr = list(mTypeMirrorsEntry.get().split(";"))
        mirrornstr = list(nTypeMirrorsEntry.get().split(";"))
        
        rowAndColumnLabel.destroy()
        rowAndColumnEntry.destroy()
        mAndnMirrorsLabel.destroy()
        mAndnMirrorsEntry.destroy()
        mTypeMirrors.destroy()
        mTypeMirrorsEntry.destroy()
        nTypeMirrors.destroy()
        nTypeMirrorsEntry.destroy()
        submitBtn.destroy()
        welcomeMessage.destroy()
        safebetImage.destroy()
         
       
        safe_bet(r,c,m,n,mirrormstr, mirrornstr)
            
        exitButton = Button(con_exit_frame, image = exitbuttonImg,border = 0, bg =  "#1b0b54",command = root.destroy).pack() 
        
    
    
    
    safebetImg = ImageTk.PhotoImage(Image.open("safebetimg.jpg"))
    safebetImage = Label(root,image = safebetImg, bg = "#1b0b54")
    safebetImage.image = safebetImg
    safebetImage.pack()
    
    
    welcomeMsgImg = ImageTk.PhotoImage(Image.open("welcomemessage.jpg"))
    welcomeMessage = Label(root,image = welcomeMsgImg, bg = "#1b0b54")
    welcomeMessage.image = welcomeMsgImg
    welcomeMessage.place(x = 170,y = 100)
    
  
    rowAndColumnLabel = Label(root,text = "Enter The dimensions of your Safe(rowsxcolumns)",font=("Alegreya", 13), fg =  "#ffffff", bg = "#1b0b54")
    rowAndColumnLabel.place(x = 120,y = 260)
    rowAndColumnEntry = Entry(width =40)
    rowAndColumnEntry.place( x = 125, y = 290)
    
    
    
    mAndnMirrorsLabel = Label(root,text = "Enter The total number of M and N type mirrors(M,N)",font=("Alegreya", 13), fg =  "#ffffff", bg = "#1b0b54")
    mAndnMirrorsLabel.place(x = 520, y = 260)
    mAndnMirrorsEntry = Entry(width = 40)
    mAndnMirrorsEntry.place(x = 525, y = 290)
    
    mTypeMirrors = Label(root,text = "Enter The positions of M type mirrors([r1,c1];[r1 c2])",font=("Alegreya", 13), fg =  "#ffffff", bg = "#1b0b54")
    mTypeMirrors.place(x = 120,y = 360)
    mTypeMirrorsEntry = Entry(width = 40)
    mTypeMirrorsEntry.place( x = 125, y = 390)
    
    nTypeMirrors = Label(root,text = "Enter The positions of N type mirrors([r1,c1];[r1 c2])",font=("Alegreya", 13), fg =  "#ffffff", bg = "#1b0b54")
    nTypeMirrors.place(x = 520, y = 360)
    nTypeMirrorsEntry = Entry(width = 40)
    nTypeMirrorsEntry.place( x = 525, y = 390)
    
    submitBtn = Button(root, image = donebuttonImg,border = 0, bg =  "#1b0b54", command = thirdWindow )
    submitBtn.pack(padx = 10, pady = 10,side = BOTTOM) 


    

main_frame = Frame(root, bg = "#1b0b54",highlightbackground= "#1b0b54",highlightthickness = 0, bd = 0 )
main_frame.pack()



main_canvas = Canvas(main_frame, width = 800, height = 500,bd = 0,bg = "#1b0b54",highlightbackground= "#1b0b54",highlightthickness = 0)
main_canvas.pack(fill = "both", expand = True)

main_canvas.create_image(0,0,image = img, anchor = "nw")



continue_btn =  Button(main_frame,image =continuebuttonImg,  bg = "#1b0b54", borderwidth = 0, command = secondWindow)
continue_btn.pack(padx = 10, pady = 10,side  = BOTTOM)


root.mainloop()
```
